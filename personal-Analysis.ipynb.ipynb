```
import pandas as pd

expenses = {
    "Date": ["2026-06-01","2026-06-02","2026-06-03","2026-06-04"],
    "Category": ["Food","Travel","Shopping","Bills"],
    "Amount": [500,300,1000,700]
}

df = pd.DataFrame(expenses)

print(df)
```

```
total = df["Amount"].sum()

print("Total Expense =", total)
```

```
category_expense = df.groupby("Category")["Amount"].sum()

print(category_expense)
```

```
highest = category_expense.idxmax()

print("Highest Spending Category:", highest)
```

```
import matplotlib.pyplot as plt

category_expense.plot(kind="bar")

plt.title("Category Wise Expenses")
plt.xlabel("Category")
plt.ylabel("Amount")

plt.show()
```

```
category_expense.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Expense Distribution")

plt.show()
```

```
print("----- PERSONAL EXPENSE REPORT -----")

print("Total Expense:", total)

print("Highest Spending Category:", highest)

print("\nCategory Wise Expense:")
print(category_expense)
```
